```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def log(x):
    return (1/(1+np.exp(-x)))
```

```python
x = np.arange(-20,20,0.01)
y = log(x)
```

```python
plt.plot(x, y)
```

```python
def bin_cross_entrop(y, label):
    return(-(label*np.log(y) + (1-label)*np.log(1-y)))
```

```python
loss_1 = bin_cross_entrop(y, 1)
loss_0 = bin_cross_entrop(y, 0)
```

```python
plt.figure(figsize=(12,10))
plt.plot(y, loss_1, linewidth=3, label='label 1')
plt.plot(y, loss_0, linewidth=3, label='label 0')
plt.legend(loc='upper left', borderpad=1, fontsize=18)
plt.show()
```

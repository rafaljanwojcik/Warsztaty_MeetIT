```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
import time

%matplotlib inline
```

```python
def quadratic(x):
    return(x**2 + 5*x + 6)
```

```python
def linear(x):
    return(2*x + 6)
```

```python
x = np.arange(-30,31,1)
#y = linear(x)
y = quadratic(x)
```

```python
plt.figure(figsize=(10,8))
plt.plot(x, y)
```

Analytically - (-b/2a, -delta/4a)

```python
def derivative(fun, x, delta):
    return((fun(x+delta) - fun(x))/delta)
```

```python
def grad_descent(fun, start_point, alpha, derivative, delta, iters):
    y = fun(x)
    for i in range(0,iters):
        if i == 0:
            current_x = start_point
            
        grad = derivative(fun, current_x, delta)
        history.append([current_x, grad])
        
        current_x = current_x - alpha*grad
        
        x_grad= np.array([i[0] for i in history])
        y_grad = fun(x_grad)
        
        clear_output(wait=True)
        plt.figure(figsize=(16,8))
        plt.plot(x, y)
        plt.scatter(x_grad, y_grad, color='r')
        plt.show()
        time.sleep(0.5)    #sleep
```

```python
#learning rates: 
#linear: ok - 1, too small - 0.1
#quadratic: ok - 0.1, too big, 1.02, too small - 0.01
#np.exp: ok - 1e-13
#np.log: ok - 150
start = x[-1]
history = []
grad_descent(quadratic, start, 0.1, derivative, 1e-6, 20)
```
